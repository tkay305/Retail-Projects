# Exploratory Analysis


## What is the distribution of Customers by Sales Method?

- Of the 13,924 customers analysed, half were contacted by Email, under a third by Call and the remaining sixth were contacted by both Email and Call.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns


customer_by_method = df.groupby(by="sales_method").agg({'customer_id':'count'}).reset_index()
pal = sns.color_palette("colorblind", customer_by_method["sales_method"].nunique())
rank = customer_by_method["sales_method"].argsort()
sns.barplot( y= "customer_id", x= "sales_method", palette=np.array(pal[::-1])[rank],data= customer_by_method).set(title = 'Customers by Sales Method')
plt.xlabel('Sales Method')
plt.ylabel('Number of Customers')
plt.savefig('customers_by_sales_method.png')
plt.show()

# were an equal number of initial customers approached? from data frame, all we can see is there is a higher number of customers that purchased new products when contacted by email. What about revenue per customer by type?
```

```python
customer_by_method["perc_total_customers"] = customer_by_method["customer_id"]/customer_by_method["customer_id"].sum()
customer_by_method
```

## What is the distribution of Revenue by Sales Method?

51% of all revenue from new product sales were earned from customers contacted by Email, 31% of revenue was from customers contacted by Email and Call. The customers contacted by phone Call contributed 17% of new product sales. Despite having a smaller customer base, the group that was contacted by Email and Call earned almost double the revenue of the group that was contacted by only calls.
There is room to earn higher revenue by contacting more customers by Email + Call rather than through Call only.

```python
revenue_by_method = df.groupby(by="sales_method").agg({'revenue':'sum'}).reset_index()
sns.barplot( y= "revenue", x= "sales_method", palette=np.array(pal[::-1])[rank], data= revenue_by_method).set(title = 'Revenue by Sales Method')
plt.xlabel('Sales Method')
plt.ylabel('Revenue Earned ($)')
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.savefig('revenue_by_sales_method.png')
plt.show()


```

### Revenue by State

```python
customer_by_state = df.groupby(by="state").agg({'revenue':'sum','customer_id':'count','years_as_customer':'median'}).reset_index().sort_values(by='customer_id',ascending=False)
customer_by_state["perc_total_customers"] = customer_by_state["customer_id"]/customer_by_state["customer_id"].sum()
customer_by_state["perc_total_revenue"] = customer_by_state["revenue"]/customer_by_state["revenue"].sum()
customer_by_state = customer_by_state.sort_values(by="customer_id", ascending=False).head(10)
pal_4 = sns.color_palette("colorblind", customer_by_state["perc_total_customers"].nunique())
rank_4 = customer_by_state["perc_total_customers"].argsort()
sns.barplot( x= "perc_total_customers", y= "state",palette=np.array(pal_4[::-1])[rank_4], data= customer_by_state).set(title = 'Percentage of Total Customers by State - Top 10 States')
plt.xlabel('Percentage of Total Customers')
plt.ylabel('State')
plt.savefig('customer_by_state.png')
plt.show()
```

## Revenue over Time by Sales Method

### Email campaign structure:
_Week 1: First email sent out_
_Week 4: Second email sent out_

- The first week of the Email campaign earned the highest revenue over the six week period, across all sales methods. The two weeks following, revenue declined, down to half of week 1's earnings by the third week. 
- In week 4, when second emails were sent out, there was a slight increase in revenue, before it declined again in weeks 5 and 6.
  
### Call campaign structure: 
_One 30 minute call (no follow up or call to action recorded) in week 1_

- The call method has the lowest revenue earned over time across all methods, though revenue earnings peak in weeks 4 and 5.

### Call + Email campaign structure 
_Email in week 1, 10 minute follow up call in week 2_

- For this method, we see revenue follow an increasing pattern from weeks 1 to 5, and a slight decline in revenues earned in the final week.

 Given that the Call + Email group **had a third of the customers seen in the Email only group, and a higher revenue earned per customer**, it appears that the Email + Call method has the highest revenue earning potential if exploited.



```python
revenue_by_method_week = df.groupby(by=["sales_method" , "week"]).agg({'revenue':'sum'}).reset_index()
revenue_by_method_week
sns.barplot( y= "revenue", x= "week", hue= "sales_method", data= revenue_by_method_week).set(title = 'Revenue earned per week')
plt.legend(title="Sales Method")
plt.xlabel('Week of Campaign')
plt.ylabel('Revenue Earned ($)')
plt.gca().yaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.savefig('revenue_per_week')
plt.show()
```

# Business Metric Recommendations

The goals of the sales team is to have insights into approaches that will increase new product sales. We can measure this by tracking revenue earned against a specific dimension.

A good way to track whether sales are effective is analysing the areas with the most potential, namely:

- **Customers by State** : The data shows that currently, new product sales are concentrated in 10 major states. Across the dimensions of years as customer, revenue per customer and sales units - there appears to be no major differences between the different states. The top performing states have the highest number of customers and in turn the highest revenue.

- By tracking the number of customers that purchased new products in each state, we would have visibility into the increase in product reach at state-level. Currently, the bottom 20 states could expand customer numbers to reach at least 250.

```python
avg_revenue_by_state = df.groupby(by="state").agg({'customer_id':'count'}).reset_index().sort_values(by='customer_id',ascending=False).head(20)

pal_3 = sns.color_palette("colorblind", avg_revenue_by_state["customer_id"].nunique())
rank_3 = avg_revenue_by_state["customer_id"].argsort()
sns.barplot( x= "customer_id", y= "state", palette=np.array(pal_3[::-1])[rank_3], data= avg_revenue_by_state).set(title = 'Number new product Customers - Top 20 States')
plt.xlabel('Number of Customers - new products only')
plt.ylabel('State')
plt.gca().xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```

```python
bottom_customers_by_state = df.groupby(by="state").agg({'customer_id':'count'}).reset_index().sort_values(by='customer_id',ascending=False).tail(20)

pal_4 = sns.color_palette("colorblind", bottom_customers_by_state["customer_id"].nunique())
rank_4 = bottom_customers_by_state["customer_id"].argsort()
sns.barplot( x= "customer_id", y= "state", palette=np.array(pal_3[::-1])[rank_3], data= bottom_customers_by_state).set(title = 'Number new product Customers - Bottom 20 States')
plt.xlabel('Number of Customers - new products only')
plt.ylabel('State')
plt.gca().xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```
